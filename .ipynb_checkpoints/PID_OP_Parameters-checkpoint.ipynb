Si chequeás plot_all, plotea el parametro para todos los vds disponibles, en función de un gm/id calculado para el vds que indica el slider. 
Sino, plotea el parametro para un solo vds en función de gm/Id calculado en ese vds.

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Apr  8 22:42:22 2020

@author: matias
"""

CBD = 'CBD'
CBS = 'CBS'
CGB = 'CGB'
CGD = 'CGD'
CGS = 'CGS'
IDS = 'IDS'
SFL = 'SFL'
SID = 'SID'
GDS = 'GDS'
GM = 'GM'
VTO = 'VTO'

import numpy as np
import matplotlib.pyplot as plt 
import ipywidgets as widgets 
from ipywidgets import interact, interact_manual
%matplotlib inline


file = open('./op_parameters.txt', 'r')
buffer = file.read()
parameters_list = buffer.split('\n') #Lista de strings
steps_vds = int(parameters_list[0].strip(";"))
W = float(parameters_list[1].strip(";"))
L = float(parameters_list[2].strip(";"))
#vds = np.arange(0, 1.62, 1.6/float(steps_vds))
steps_vgs = 1600
data_list = []
new_list = []
#22 == Numero de parametros + su respectivo vgs
#Despues agregar los valores en el script de ocean.

for index in range(0, len(parameters_list) - 1, 1):
    if parameters_list[index].startswith(";") == False:
        data_list.append(parameters_list[index])

parameters_list = []

for index in range(0, len(data_list) - 1, 1):
    parameters_list.append(data_list[index].split(","))
    
data_list = []

for g_index in range(0, steps_vds, 1):
#Creo una lista de los CSV

    data_list.append(parameters_list[1601*g_index:1601*(g_index+1)])

data_list = np.array(data_list).astype(float)
parameters_list = []    

    
#Vectorizo cada parametro
vgs = data_list[0, :, 0] 
cbd = data_list[:, :, 1]
cbs = data_list[:, :, 3]
cgb = data_list[:, :, 5]
cgd = data_list[:, :, 7]
cgs = data_list[:, :, 9]
ids = data_list[:, :, 11]
sfl = data_list[:, :, 13]
sid = data_list[:, :, 15]
gds = data_list[:, :, 17]
gm = data_list[:, :, 19]
vto = data_list[:, :, 21]


    

def select_parameter(param = 'CBD', vds = 0.5, plot_all = False):
    
    x = gm[int(vds*steps_vds/1.6), :]/ids[int(vds*steps_vds/1.6), :]
    if(plot_all == 0):     
        
        if(param == 'CBD'):
            plt.plot(x,cbd[int(vds*steps_vds/1.6), :])

        if(param == 'CBS'):
            plt.plot(x,cbs[int(vds*steps_vds/1.6), :])

        if(param == 'CGB'):
            plt.plot(x,cgb[int(vds*steps_vds/1.6), :])

        if(param == 'CGD'):
            plt.plot(x,cgd[int(vds*steps_vds/1.6), :])

        if(param == 'CGS'):
            plt.plot(x,cgs[int(vds*steps_vds/1.6), :])

        if(param == 'IDS'):
            plt.plot(x,ids[int(vds*steps_vds/1.6), :])

        if(param == 'SFL'):
            plt.plot(x,sfl[int(vds*steps_vds/1.6), :])

        if(param == 'SID'):
            plt.plot(x,sid[int(vds*steps_vds/1.6), :])

        if(param == 'GDS/ID'):
            plt.plot(x,gds[int(vds*steps_vds/1.6), :]/ids[int(vds*steps_vds/1.6), :])

        if(param == 'GM'):
            plt.plot(x, gm[int(vds*steps_vds/1.6), :])

        if(param == 'VTO'):
            plt.plot(x,vto[int(vds*steps_vds/1.6), :])
        if(param == "In"):
            plt.plot(x, ids[int(vds*steps_vds/1.6), :]/(W/L))           
    else:
        if(param == 'CBD'):
            for index in range(1, steps_vds, 1):
                plt.plot(x,cbd[index, :])

        if(param == 'CBS'):
            for index in range(1, steps_vds, 1):
                plt.plot(x,cbs[index, :])

        if(param == 'CGB'):
            for index in range(1, steps_vds, 1):
                plt.plot(x,cgb[index, :])

        if(param == 'CGD'):
            for index in range(1, steps_vds, 1):
                plt.plot(x,cgd[index, :])

        if(param == 'CGS'):
            for index in range(1, steps_vds, 1):
                plt.plot(x,cgs[index, :])

        if(param == 'IDS'):
            for index in range(1, steps_vds, 1):
                plt.plot(x,ids[index, :])

        if(param == 'SFL'):
            for index in range(1, steps_vds, 1):
                plt.plot(x,sfl[index, :])

        if(param == 'SID'):
            for index in range(1, steps_vds, 1):
                plt.plot(x,sid[index, :])

        if(param == 'GDS/ID'):
            for index in range(1, steps_vds, 1):
                plt.plot(x,gds[index, :]/ids[index, :])


        if(param == 'GM'):
            for index in range(1, steps_vds, 1):
                plt.plot(x, gm[index, :])

        if(param == 'VTO'):
            for index in range(1, steps_vds, 1):
                plt.plot(x,vto[index, :])
                
        if(param == "In"):
            for index in range(1, steps_vds, 1):
                plt.plot(x, ids[index, :]/(W/L))


plt.show()
# ;cbd;cbs;cgb;cgd;cgs;ids;sfl;sid;gds;gm;vto
widgets.interact(select_parameter, param = ["CBD", "CBS", "CGB", "CGD", "CGS", "IDS", "SFL", "SID", "GDS/ID", "GM", "VTO", "In"], vds = (0.02, 1.6*(1 - 1/steps_vds), 1.6/steps_vds), plot_all = False)
```
